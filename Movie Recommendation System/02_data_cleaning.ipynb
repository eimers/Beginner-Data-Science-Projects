# Movie Recommendation System — 02 Preprocessing & Split

We restrict to active users and popular movies (a denser, tractable matrix), build the user × movie rating matrix, and split known ratings into train/test so we can measure how well each model predicts held-out ratings.

```python
import numpy as np, pandas as pd
%matplotlib inline
import utils

ratings = utils.load_ratings()
print('raw ratings:', ratings.shape)
```

## 1. Filter to active users / popular movies

```python
df = utils.filter_popular(ratings, max_users=610, max_items=2000, min_user=5, min_item=5)
print('after filtering:', df.shape)
print('users:', df.user.nunique(), '| movies:', df.item.nunique())
```

## 2. Train/test split of ratings

A random 80/20 hold-out of individual ratings; test pairs whose user or movie is unseen in train are moved back to train.

```python
train, test = utils.train_test_split_ratings(df, test_size=0.2, seed=42)
print('train ratings:', len(train), '| test ratings:', len(test))
```

## 3. Build the user-item matrix

```python
M = utils.build_user_item(train)
print('matrix shape:', M.shape)
print('sparsity: %.3f' % utils.sparsity(M))
M.iloc[:5,:6]
```

## 4. Save processed splits

```python
train.to_csv('data/train.csv', index=False)
test.to_csv('data/test.csv', index=False)
print('saved data/train.csv', train.shape, 'and data/test.csv', test.shape)
```
