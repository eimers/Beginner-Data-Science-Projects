# Book Recommendation System — 02 Preprocessing & Split

Restrict to active users / popular books, build the rating matrix, split ratings into train/test.

```python
import pandas as pd
import utils
ratings=utils.load_ratings()
print('raw:',ratings.shape)
```

## 1. Filter to active users / popular books

```python
df=utils.filter_popular(ratings,max_users=2000,max_items=2000,min_user=5,min_item=5)
print('after filter:',df.shape,'| users',df.user.nunique(),'| books',df.item.nunique())
```

## 2. Train/test split

```python
train,test=utils.train_test_split_ratings(df,test_size=0.2,seed=42)
print('train',len(train),'| test',len(test))
```

## 3. Matrix + save

```python
M=utils.build_user_item(train)
print('matrix',M.shape,'sparsity %.3f'%utils.sparsity(M))
train.to_csv('data/train.csv',index=False); test.to_csv('data/test.csv',index=False)
print('saved train/test')
```
