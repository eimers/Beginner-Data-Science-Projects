# A/B Test Analysis — 01 The Experiment

We **simulate** a controlled A/B test with a *known* true effect (control 11.0% vs treatment 12.8% conversion). Simulating lets us later verify the statistical test actually recovers the ground-truth uplift — the honest way to learn A/B analysis.

```python
import numpy as np, pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import utils
A,B=utils.simulate_experiment()
print('control n=%d conv=%d (%.4f)'%(len(A),A.sum(),A.mean()))
print('treatment n=%d conv=%d (%.4f)'%(len(B),B.sum(),B.mean()))
print('true effect built in: +0.018 absolute')
```

## 1. Observed conversion rates

```python
fig,ax=plt.subplots(figsize=(6,3.5)); ax.bar(['Control (A)','Treatment (B)'],[A.mean(),B.mean()],color=['#4c72b0','#55a868']); ax.set_ylabel('conversion rate'); ax.set_title('Observed conversion by group'); 
for i,v in enumerate([A.mean(),B.mean()]): ax.text(i,v+0.002,f'{v:.3f}',ha='center')
plt.tight_layout(); plt.show()
```

## 2. Summary

- 12,000 users per group; treatment converts higher in the sample.
- The question notebook 02 answers: **is this difference statistically significant, or could it be noise?**
