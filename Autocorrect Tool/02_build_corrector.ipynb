# Autocorrect Tool — 02 Building the Corrector

We implement Peter Norvig's classic spelling corrector. For a misspelled word we generate all strings within **1 or 2 edits** (delete, transpose, replace, insert), keep those that are real words, and return the **most frequent** candidate — `argmax P(candidate)`.

```python
import utils
freq=utils.load_frequencies()
sc=utils.SpellCorrector(freq)
print('corrector ready over',len(freq),'words')
```

## 1. The edit-distance neighbourhood

How many 1-edit and 2-edit candidates a word generates.

```python
w='korrect'
e1=sc.edits1(w); e2=sc.edits2(w)
print(f'"{w}": {len(e1):,} strings at edit-distance 1, {len(e2):,} at distance 2')
print('known 1-edit candidates:', sc.known(e1))
```

## 2. Correcting single words

```python
for w in ['speling','korrect','recieve','seperate','definately','goverment','wich','thru']:
    print(f'{w:12s} -> {sc.correct(w)}')
```

## 3. Correcting a sentence

```python
print(sc.correct_sentence('ths is a tst of the speling korrector'))
print(sc.correct_sentence('i recieved a seperate mesage yesterday'))
```

## 4. How candidates are ranked

Among known corrections, the one with the highest corpus probability wins — that is why frequency matters.

```python
w='thre'
cands=sc.known(sc.edits1(w))
import pandas as pd
print(pd.Series({c:sc.probability(c) for c in cands}).sort_values(ascending=False).head(6).to_string())
print('\nchosen:',sc.correct(w))
```
