# Keyword Extraction — 02 Three Methods

- **TF-IDF** — terms with high weight in this article relative to the rest.
- **RAKE** (Rapid Automatic Keyword Extraction) — candidate phrases split on stopwords, scored by word degree/frequency; returns multi-word **phrases**.
- **TextRank** — PageRank over a word co-occurrence graph; the most 'central' words.

```python
import utils, warnings; warnings.filterwarnings('ignore')
df=utils.load_articles()
article=df.iloc[0]['article']
print('reference summary:',df.iloc[0]['summary'][:200])
```

## 1. TF-IDF keywords

```python
print(utils.keywords_tfidf(article,10))
```

## 2. RAKE keyphrases

```python
for kw in utils.keywords_rake(article,8): print(' -',kw)
```

## 3. TextRank keywords

```python
print(utils.keywords_textrank(article,10))
```
