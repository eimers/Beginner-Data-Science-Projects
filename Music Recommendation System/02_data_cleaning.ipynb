# Music Recommendation System — 02 Preprocessing & Split

Build the implicit play matrix (log-compressed) and hold out a fraction of each user's artists for ranking evaluation.

```python
import pandas as pd
import utils
plays=utils.load_interactions()
print('raw:',plays.shape)
```

## 1. Filter to active users / popular artists

```python
df=utils.filter_popular(plays,max_users=1892,max_items=2000,min_user=5,min_item=5)
print('after filter:',df.shape,'| users',df.user.nunique(),'| artists',df.item.nunique())
```

## 2. Per-user train/test split (hold out 20% of each user's artists)

```python
train,test=utils.train_test_split_interactions(df,test_size=0.2,seed=42)
print('train',len(train),'| test',len(test))
```

## 3. Build log-compressed matrix + save

```python
M=utils.build_matrix(train,log=True)
print('matrix',M.shape,'sparsity %.3f'%utils.sparsity(M))
train.to_csv('data/train.csv',index=False); test.to_csv('data/test.csv',index=False)
print('saved train/test')
```
