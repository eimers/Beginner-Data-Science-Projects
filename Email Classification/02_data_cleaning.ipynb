# Email Classification — 02 Text Cleaning

Produce a `clean_text` column: lowercase, strip URLs/HTML/punctuation/digits, remove stopwords and very short tokens. This is the text the TF-IDF vectorizer consumes in notebook 03.

```python
import pandas as pd
import utils
df=utils.load_data()
print('rows:',len(df))
```

## 1. Apply cleaning

```python
df['clean_text']=df['text'].astype(str).map(utils.clean_text)
df=df[df['clean_text'].str.strip()!='']
print('after cleaning:',len(df),'rows')
```

## 2. Before / after example

```python
ex=df.iloc[0]
print('RAW  :',str(ex['text'])[:200].replace(chr(10),' '))
print('\nCLEAN:',ex['clean_text'][:200])
```

## 3. Save cleaned data

```python
df[['clean_text','label']].to_csv('data/emails_cleaned.csv',index=False)
print('saved data/emails_cleaned.csv',df.shape)
```
