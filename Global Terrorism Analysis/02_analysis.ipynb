# Global Terrorism Analysis — 02 Analysis

Where and when attacks concentrate, how tactics break down, and who the main groups are.

```python
import pandas as pd, numpy as np
import matplotlib.pyplot as plt, seaborn as sns
import warnings; warnings.filterwarnings('ignore')
%matplotlib inline
import utils
df=utils.load_data()
print('attacks',len(df),'years',int(df.iyear.min()),'-',int(df.iyear.max()))
```

## 1. Attacks over time

```python
yr=df['iyear'].value_counts().sort_index()
fig,ax=plt.subplots(figsize=(13,3.5)); ax.plot(yr.index,yr.values,color='#c44e52'); ax.fill_between(yr.index,yr.values,alpha=0.3,color='#c44e52'); ax.set_title('Terrorist attacks per year'); plt.tight_layout(); plt.show()
print('peak year:',int(yr.idxmax()),'=',int(yr.max()),'attacks')
```

## 2. Most-affected countries & regions

```python
fig,ax=plt.subplots(1,2,figsize=(14,3.8))
df['country_txt'].value_counts().head(10).plot(kind='barh',ax=ax[0],color='#4c72b0'); ax[0].invert_yaxis(); ax[0].set_title('Top countries')
df['region_txt'].value_counts().head(8).plot(kind='barh',ax=ax[1],color='#dd8452'); ax[1].invert_yaxis(); ax[1].set_title('Top regions')
plt.tight_layout(); plt.show()
```

## 3. Attack types & target types

```python
fig,ax=plt.subplots(1,2,figsize=(14,3.8))
df['attacktype1_txt'].value_counts().head(8).plot(kind='barh',ax=ax[0],color='#8172b3'); ax[0].invert_yaxis(); ax[0].set_title('Attack types')
df['targtype1_txt'].value_counts().head(8).plot(kind='barh',ax=ax[1],color='#55a868'); ax[1].invert_yaxis(); ax[1].set_title('Target types')
plt.tight_layout(); plt.show()
```

## 4. Top perpetrator groups (named)

```python
g=df['gname'].value_counts().drop('Unknown',errors='ignore').head(10)
fig,ax=plt.subplots(figsize=(10,3.5)); g.plot(kind='barh',ax=ax,color='#c9a227'); ax.invert_yaxis(); ax.set_title('Most active named groups'); plt.tight_layout(); plt.show()
print('attack success rate: %.1f%%'%(df['success'].mean()*100))
```

## 5. Key findings

- **181,691 attacks, 1970–2017**, with a dramatic **spike after 2011 peaking in 2014 (16,903 attacks)** — the ISIL/Taliban surge era.
- **Concentrated in the Middle East/North Africa and South Asia** — Iraq (24,636), Pakistan (14,368), Afghanistan (12,731), India, Colombia lead.
- **Bombings/explosions are the dominant tactic (~49%)**, then armed assault and assassination.
- **Taliban, ISIL, and Shining Path** are the most active named groups (though a large share of attacks are attributed to 'Unknown').
- **89% of attacks 'succeeded'** by the database's definition (the attack occurred as intended) — a sobering base rate.
- Note: this GTD extract excludes casualty counts, so the analysis focuses on frequency, geography, tactics, and actors.
