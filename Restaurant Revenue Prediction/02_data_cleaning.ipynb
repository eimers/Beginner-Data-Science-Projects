# Restaurant Revenue Prediction — 02 Data Cleaning & Feature Engineering

The data has no missing values, so cleaning is light. The work here is **feature engineering**: turning `Open Date` into restaurant age, dropping the high-cardinality `City` column, and one-hot encoding the low-cardinality categoricals.

```python
import numpy as np, pandas as pd
%matplotlib inline
import utils

df = utils.load_data()
print('raw shape:', df.shape)
```

## 1. Quality checks

```python
print('missing values total:', df.isnull().sum().sum())
print('duplicate rows:', df.duplicated().sum())
print('City cardinality:', df['City'].nunique(), '(too granular for 137 rows -> dropped)')
```

## 2. Feature engineering

`create_features` parses `Open Date` → `restaurant_age_years` (from a 2015-01-01 snapshot) and drops `Id`, `Open Date`, and `City`.

```python
fe = utils.create_features(df)
print('after feature engineering:', fe.shape)
print('new column:', 'restaurant_age_years' in fe.columns)
fe[['restaurant_age_years', 'City Group', 'Type', 'revenue']].head()
```

## 3. One-hot encode categoricals

`City Group` (2 levels) and `Type` (3 levels) are one-hot encoded with `drop_first=True` to avoid the dummy-variable trap.

```python
X, y, feats = utils.preprocess_data(df)
print('feature matrix:', X.shape)
print('encoded categorical columns:', [c for c in X.columns if c.startswith(('City Group_', 'Type_'))])
print('total features:', len(feats))
```

## 4. Save cleaned dataset

```python
cleaned = X.copy()
cleaned['revenue'] = y.values
cleaned.to_csv('data/train_cleaned.csv', index=False)
print('saved data/train_cleaned.csv', cleaned.shape)
print('no missing:', cleaned.isnull().sum().sum() == 0)
```
