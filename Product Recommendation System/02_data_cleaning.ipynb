# Product Recommendation System — 02 Preprocessing & Split

Aggregate purchases into a customer × product implicit matrix (log-compressed quantities) and hold out a fraction of each customer's products for ranking evaluation.

```python
import pandas as pd
import utils
buys=utils.load_interactions()
print('raw pairs:',buys.shape)
```

## 1. Filter to active customers / popular products

```python
df=utils.filter_popular(buys,max_users=2000,max_items=800,min_user=5,min_item=5)
print('after filter:',df.shape,'| customers',df.user.nunique(),'| products',df.item.nunique())
```

## 2. Per-customer train/test split (hold out 20% of products)

```python
train,test=utils.train_test_split_interactions(df,test_size=0.2,seed=42)
print('train',len(train),'| test',len(test))
```

## 3. Build log-compressed matrix + save

```python
M=utils.build_matrix(train,log=True)
print('matrix',M.shape,'sparsity %.3f'%utils.sparsity(M))
train.to_csv('data/train.csv',index=False); test.to_csv('data/test.csv',index=False)
print('saved train/test')
```
