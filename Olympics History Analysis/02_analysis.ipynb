# Olympics History Analysis — 02 Analysis

Medal powerhouses, the rise of female participation, and how athlete bodies changed.

```python
import pandas as pd, numpy as np
import matplotlib.pyplot as plt, seaborn as sns
import warnings; warnings.filterwarnings('ignore')
%matplotlib inline
import utils
df=utils.load_data()
print('records',len(df),'athletes',df.ID.nunique())
```

## 1. Medal-winning nations (all-time)

```python
m=df[df.Medal.notna()]['NOC'].value_counts().head(10)
fig,ax=plt.subplots(figsize=(10,3.5)); m.plot(kind='bar',ax=ax,color='#c9a227'); ax.set_title('Top 10 NOCs by medals won (athlete-events)'); plt.tight_layout(); plt.show()
print(m.to_dict())
```

## 2. Rise of female participation

```python
fem=df[df.Season=='Summer'].groupby('Year')['Sex'].apply(lambda s:(s=='F').mean()*100)
fig,ax=plt.subplots(figsize=(11,3.5)); fem.plot(ax=ax,marker='o',color='#c44e52'); ax.set_title('Female share of Summer Olympians (%)'); ax.set_ylabel('% female'); plt.tight_layout(); plt.show()
print('1936: %.1f%%   2016: %.1f%%'%(fem.get(1936,float('nan')),fem.get(2016,float('nan'))))
```

## 3. Athletes are getting bigger — height over time

```python
ht=df.groupby('Year')['Height'].mean()
fig,ax=plt.subplots(figsize=(11,3.5)); ht.plot(ax=ax,marker='.',color='#4c72b0'); ax.set_title('Mean athlete height by year (cm)'); plt.tight_layout(); plt.show()
```

## 4. Most-contested sports

```python
sp=df['Sport'].value_counts().head(10)
fig,ax=plt.subplots(figsize=(10,3.5)); sp.plot(kind='barh',ax=ax,color='#55a868'); ax.invert_yaxis(); ax.set_title('Most athlete-events by sport'); plt.tight_layout(); plt.show()
```

## 5. Key findings

- **271,116 records, 135,571 unique athletes, 1896→2016.**
- **Medal powerhouses**: USA (5,637 medal-events) far ahead, then the USSR (2,503), Germany, Great Britain, France.
- **Female participation exploded**: 7.4% of Summer Olympians in 1936 → 45.5% in 2016 — one of the clearest equality trends in sport.
- **Athletes trend taller/heavier** over the century (professionalisation + specialisation).
- Athletics, Gymnastics and Swimming dominate by event volume.
