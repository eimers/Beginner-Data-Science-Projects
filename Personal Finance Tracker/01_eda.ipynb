# Personal Finance Tracker — 01 EDA

A year of personal transactions (income, discretionary spending, rent). We track where the money goes. *Data is a seeded simulation of realistic categorised transactions.*

```python
import pandas as pd, numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import utils
df=utils.load_transactions()
print('transactions:',len(df))
print('total spend $%.0f | total income $%.0f'%(-df[df.amount<0].amount.sum(),df[df.amount>0].amount.sum()))
df.head()
```

## 1. Spending by category

```python
ct=utils.category_totals(df)
fig,ax=plt.subplots(figsize=(9,3.5)); ct.plot(kind='bar',ax=ax,color='#c44e52'); ax.set_title('Total spend by category ($)'); plt.xticks(rotation=45,ha='right'); plt.tight_layout(); plt.show()
print(ct.round(0).to_dict())
```

## 2. Monthly spending

```python
ms=utils.monthly_spend(df)
fig,ax=plt.subplots(figsize=(11,3.5)); ms.plot(kind='bar',ax=ax,color='#4c72b0'); ax.set_title('Monthly spend ($)'); plt.tight_layout(); plt.show()
print('monthly spend: mean $%.0f std $%.0f'%(ms.mean(),ms.std()))
```

## 3. Summary

- ~770 transactions over the year; **rent is the single biggest line**, then discretionary categories (shopping, groceries, dining).
- Monthly spend is fairly stable (mean ~\$4,600). Notebook 02 auto-categorises transactions and looks at predictability.
