# Text Summarization — 02 Building Summarizers

Three extractive methods:

- **Lead-3** — take the first 3 sentences (a famously strong baseline for news).
- **TF-IDF** — score each sentence by the summed TF-IDF weight of its words; take the top 3.
- **TextRank** — PageRank over a sentence-similarity graph; take the 3 most central sentences.

```python
import utils
df=utils.load_articles()
article=df.iloc[0]['article']
reference=df.iloc[0]['summary']
print('reference:\n',reference)
```

## 1. Lead-3 baseline

```python
print(utils.summarize_lead(article,n=3))
```

## 2. TF-IDF sentence scoring

```python
print(utils.summarize_tfidf(article,n=3))
```

## 3. TextRank

```python
print(utils.summarize_textrank(article,n=3))
```

## 4. Side-by-side on another article

```python
a=df.iloc[5]['article']
print('REF     :',df.iloc[5]['summary'][:200])
print('\nLEAD-3  :',utils.summarize_lead(a)[:200])
print('\nTF-IDF  :',utils.summarize_tfidf(a)[:200])
print('\nTEXTRANK:',utils.summarize_textrank(a)[:200])
```
