# Restaurant Recommendation System — 02 Preprocessing & Split

Build the user × restaurant matrix and split ratings into train/test.

```python
import pandas as pd
import utils
ratings=utils.load_ratings()
print('raw:',ratings.shape)
```

## 1. Filter (light — data is already small)

```python
df=utils.filter_popular(ratings,max_users=200,max_items=200,min_user=3,min_item=3)
print('after filter:',df.shape,'| users',df.user.nunique(),'| places',df.item.nunique())
```

## 2. Train/test split

```python
train,test=utils.train_test_split_ratings(df,test_size=0.2,seed=42)
print('train',len(train),'| test',len(test))
```

## 3. Matrix + save

```python
M=utils.build_user_item(train)
print('matrix',M.shape,'sparsity %.3f'%utils.sparsity(M))
train.to_csv('data/train.csv',index=False); test.to_csv('data/test.csv',index=False)
print('saved train/test')
```
