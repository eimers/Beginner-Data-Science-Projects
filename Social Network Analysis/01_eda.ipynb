# Social Network Analysis — 01 EDA

A real friendship network: **SNAP Facebook** combined ego-networks (anonymised). We explore its size and degree structure before analysing centrality and communities.

```python
import numpy as np, pandas as pd
import matplotlib.pyplot as plt
import warnings; warnings.filterwarnings('ignore')
%matplotlib inline
import utils
G=utils.load_graph()
s=utils.basic_stats(G)
for k,v in s.items(): print(f'{k:16s}: {v:.4f}' if isinstance(v,float) else f'{k:16s}: {v}')
```

## 1. Degree distribution (heavy tail)

```python
degs=[d for _,d in G.degree()]
fig,ax=plt.subplots(1,2,figsize=(12,3.5))
ax[0].hist(degs,bins=60,color='#4c72b0'); ax[0].set_title('Degree distribution'); ax[0].set_xlabel('degree')
ax[1].hist(degs,bins=60,color='#c44e52'); ax[1].set_yscale('log'); ax[1].set_title('Degree (log y) — scale-free-ish tail')
plt.tight_layout(); plt.show()
print('avg degree %.1f, max %d'%(np.mean(degs),max(degs)))
```

## 2. Summary

- **4,039 people, 88,234 friendships**, one connected component.
- **High average clustering (~0.61)** — friends-of-friends are often friends (the hallmark of real social graphs).
- A heavy-tailed degree distribution (a few hub users with 1000+ friends). Notebook 02 finds the bridges and communities.
